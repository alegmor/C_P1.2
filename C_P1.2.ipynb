Una pared de concreto, que tiene un área superificial de $A=20 m^2$ y $L=0.30 m$ de espesor, separa el aire acondicionado de una habitación del aire ambiental. La temperatura de la superficie interna de la pared se mantiene a $T_1=25 ^\circ C$ y la conductividad térmica del concreto es $k=1 W/mK$.

 (a) Determine la pérdida de calor a través de la pared para temperaturas ambientes $T_a$ en el rango de $-15 ^\circ C$ a $ 38 ^\circ C$, que corresponden a extremos de invierno y verano, respectivamente. Muestre en forma gŕafica sus resultados.

 (b) Tambien grafique la pérdida de calor como función de la temperatura ambiente para materiales de la pared que tengan conductividades térmicas de $k_1=0.75$ y $k_2=1.25 W/mK$. Explique sus resultados

```python
figura1
```

```python
#Solución de (a)
k  = 1 
A  = 20
T1 = 25
L  = 0.3
Ta = np.arange(-15,38) 
qx = (k*A*(T1-Ta))/L #Ley de Fourier

plt.plot(Ta,qx)
plt.ylabel("Pérdida de calor $q_x$(W)")
plt.xlabel("Temperatura ambiente $T_a$ $(^ \circ C)$")
plt.show()
```

```python
#Solución de (b)
k   = 1
k1  = 0.75
k2  = 1.25
A   = 20
T1  = 25
L   = 0.3
Ta  = np.arange(-15,38) 
qx  = (k*A*(T1-Ta))/L
qx1 = (k1*A*(T1-Ta))/L
qx2 = (k2*A*(T1-Ta))/L

plt.plot(Ta,qx1,label="$k=0.75$ W/mK")
plt.plot(Ta,qx,label="$k=1$ W/mK")
plt.plot(Ta,qx2,label="$k=1.25$ W/mK")
#plt.axline((0, 0), (40, 0), linewidth=1,ls='--', color='gray')
#plt.axline((25, -1000), (25, 4000), linewidth=1,ls='--', color='gray')
plt.xticks(np.arange(-20,45,5))
plt.grid(which='major', alpha=0.5)
plt.ylabel("Pérdida de calor $q_x$(W)")
plt.xlabel("Temperatura ambiente $T_a$ $(^ \circ C)$")
plt.legend()
plt.show()
```
