La ecuación de calor unidimensional en estado estacionario y sin generación de calor es 

$$\frac{d}{dx} \left(k \frac{dT}{dx} \right) = 0,$$

cuando la conductividad térmica $k$  no depende de la distancia $x$ y es diferente de cero, la ecuación anterior es

$$\frac{d^2 T}{dx^2}=0.$$

La solución general de esta ecuación es

$$T(x)=C_1x+C_2.$$

Para obtener el valor de las constantes $C_1$ y $C_2$ evaluamos la condiciones de frontera 

$$T(0)=T_{1}, \qquad $$

$$T(L)=T_{2}$$

Evaluamos la condición de frontera en $x=0$ 

$$T_{1}= C_2$$.

Evaluamos en $x=L$

$$T_{2}=C_1 L+C_2 = C_1L+T_{1}, \qquad \frac{T_{1}-T_{2}}{L}=C_1$$


Sustitumos $C_1$ y $C_2$ en la ecuación (3) y encontramos la distribución de temperatura

$$T(x) = \frac{\left( T_{2}- T_{1} \right)}{L} x+ T_{1}$$

Calculamos el flujo de calor por unidad de área $q_x^{''}$ usando la ley de Fourier. El resultado es

$$q_x^{''} = - k \frac{\partial T(x)}{\partial x}= k \frac{T_1-T_2}{L} $$ 

### Problema 1.2

Una pared de concreto, que tiene un área superificial de $A=20 \mbox{m}^2$ y $L=0.30 \mbox{m}$ de espesor, separa el aire acondicionado de una habitación del aire ambiental. La temperatura de la superficie interna de la pared se mantiene a $T_1=25 ^\circ \mbox{C}$ y la conductividad térmica del concreto es $k=1 \mbox{W/mK}$ .

 (a) Determine la pérdida de calor a través de la pared para temperaturas ambientes $T_a$ en el rango de $-15 ^\circ \mbox{C}$ a $ 38 ^\circ \mbox{C}$, que corresponden a extremos de invierno y verano, respectivamente. Muestre en forma gŕafica sus resultados.

 (b) También grafique la pérdida de calor como función de la temperatura ambiente para materiales de la pared que tengan conductividades térmicas de $k_1=0.75$ y $k_2=1.25 \mbox{W/mK}$. Explique sus resultados

```python
#Convertir unidades
#Celsius a Kelvin
#Se puede hacer más sencillo
def CelsiusKelvin(C):
    K = C+273.15
    return K
CelsiusKelvin(-15)
#25-(-15), 298.15-258.15
```

```python
#Ley de Fourier
def qx(k,A,T1,Ta,L):
    return (k*A*(T1-Ta))/L 
```

```python
#Solución de (a)

k  = 1 
A  = 20
T1 = 25
L  = 0.3
Ta = np.arange(-15,38) 
qx(k,A,T1,Ta,L)

plt.plot(Ta,qx(k,A,T1,Ta,L))
plt.ylabel("Pérdida de calor $q_x$(W)")
plt.xlabel("Temperatura ambiente $T_a$ $(^ \circ C)$")
plt.show()
```

```python
#Solución de (b)

k   = 1
k1  = 0.75
k2  = 1.25
A   = 20
T1  = 25
L   = 0.3
Ta  = np.arange(-15,38) 

plt.plot(Ta,qx(k,A,T1,Ta,L),label="$k=0.75$ W/mK")
plt.plot(Ta,qx(k1,A,T1,Ta,L),label ="$k=1$ W/mK")
plt.plot(Ta,qx(k2,A,T1,Ta,L),label="$k=1.25$ W/mK")
plt.xticks(np.arange(-20,45,5))
plt.grid(which='major', alpha=0.5)
plt.ylabel("Pérdida de calor $q_x$(W)")
plt.xlabel("Temperatura ambiente $T_a$ $(^ \circ C)$")
plt.legend()
plt.show()
```
